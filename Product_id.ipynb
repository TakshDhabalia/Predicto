```python
def getProductDetails():
    
    try:
        with open("product_info.json", 'r') as jsonFile:
            data = json.load(jsonFile)
        
        schemaFields = data['schema']['fields']
        column_names = []

        dataObjects = data['data']

        for i in range(0, len(schemaFields)):
            column_names.append(schemaFields[i]['name'])

        product_df = pd.DataFrame(data=dataObjects, columns=column_names)

        return product_df

        

    except Exception as e:
        print(e)
```

```python
product_df = getProductDetails()
df_sorted = product_df.sort_values(by="Product ID", ascending=True)
df_sorted.head(20)
```

```python
productid_dup = product_df[product_df.duplicated(subset=['Product ID'],keep=False)]
productid_dup_sorted = productid_dup.sort_values(by="Product ID", ascending=True)
productid_dup_sorted
```

```python
import pandas as pd


df_sorted = product_df.sort_values(by="Product ID", ascending=True).reset_index(drop=True)

product_count = {}


modified_ids = []


for index, row in df_sorted.iterrows():
    product_id = row["Product ID"]
    
    if product_id in product_count:
        product_count[product_id] += 1
    else:
        product_count[product_id] = 0  


    modified_ids.append(f"{product_id}-{product_count[product_id]}")


df_sorted["Modified Product ID"] = modified_ids


print(df_sorted.head(20))  


productid_dup_sorted = df_sorted[df_sorted.duplicated(subset=["Product ID"], keep=False)]
productid_dup_sorted
```
